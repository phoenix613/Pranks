```python
from matplotlib.pyplot import *
from numpy import *
```

```python
f = lambda phi, q: sqrt(2/pi)* exp(-2*(q - 3*cos(phi) + 4*sin(phi))**2)
q = linspace(-7,7,2000)
phi = linspace(0,2*pi,1000)
F = []
for p in phi:
    F.append(f(p,q))
F = array(F)    
figure(figsize = (13,8))
pcolormesh(phi/pi,q,F.T)
colorbar()
xlabel(r'$\varphi, \dfrac{1}{\pi}$', size = 20)
tick_params(labelsize = 20)
ylabel(r'$Q_{\varphi}$',size = 20)
```

```python
W = lambda psi, q, p: 2/pi * exp(-2*(q - 3*cos(psi)+4*sin(psi))**2)*exp(-2*(p + 4*cos(psi) - 3*sin(psi))**2)
Q = linspace(-6,6,2000)
P = linspace(-6,6,1000)
phi = linspace(0,2*pi,1000)
Wig = []
#psi = pi
for i, psi in enumerate([0,pi/6,pi/4,pi/3,pi/2,2*pi/3,3*pi/4,5*pi/6,pi,5*pi/4,3*pi/2,7*pi/4,2*pi]):
    for p in P:
        Wig.append(W(psi, Q, p))
    Wig = array(Wig)    
    figure(figsize = (13,8))
    pcolormesh(Q,P,Wig,cmap = 'inferno')
    colorbar()
    grid()
    xlabel(r'$Q$', size = 20)
    tick_params(labelsize = 20)
    ylabel(r'$P$',size = 20)

    title(f'phi = {psi/pi}$\pi$',size = 20)
    savefig(f'/fig{i}.png')
    close('all')
    Wig = []
```
